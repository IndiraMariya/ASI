```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import os

# Read f10.7 index and time from CSV file
csv_file = "../../data/F10.7.csv"
df = pd.read_csv(csv_file)

df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df = df[['DateTime', 'Adjusted Flux']]

# Display DataFrame
# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['DateTime'], df['Adjusted Flux'], marker='o', linestyle='-')
plt.title('Adjusted Flux Over Time')
plt.xlabel('DateTime')
plt.ylabel('Adjusted Flux')
plt.grid(True)
plt.tight_layout()
plt.show()
print(df)
```
