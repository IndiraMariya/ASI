```python
import os
import pandas as pd
import numpy
from datetime import datetime

# Directory containing PNG images
image_dir1 = "../../data/images/0335-010122-010222"
image_dir2 = "../../data/images/HMIB-010122-010222"
image_dir3 = "../../data/images/HMIIC-010122-010222"

dates = []

# List PNG files in the directory
images1 = [os.path.join(image_dir1, file) for file in os.listdir(image_dir1) if file.endswith('.jpg')]
images2 = [os.path.join(image_dir2, file) for file in os.listdir(image_dir2) if file.endswith('.jpg')]
images3 = [os.path.join(image_dir3, file) for file in os.listdir(image_dir3) if file.endswith('.jpg')]

for filepath in images1:
    date = filepath[37:-13]
    format_string = "%Y%m%d_%H%M%S"
    date_object = datetime.strptime(date, format_string)
    dates.append(date_object)

for filepath in images2:
    date = filepath[37:-13]
    format_string = "%Y%m%d_%H%M%S"
    date_object = datetime.strptime(date, format_string)
    dates.append(date_object)

for filepath in images3:
    date = filepath[38:-14]
    format_string = "%Y%m%d_%H%M%S"
    date_object = datetime.strptime(date, format_string)
    dates.append(date_object)

# Remove duplicates by converting the array to a set
dates = list(set(dates))
print(dates)

# Create a DataFrame with file paths of PNG images
data = {
    'Date' : dates
    # '0335': images1,
    # 'HMIB': images2,
    # 'HMIIC': images3,
    # 'Label': [0] * len(images1)  # Example labels for classification
}

df = pd.DataFrame(data)

# # Display DataFrame
# print(df)
```
